```python
def calculate_variance_of_sample(seriesObj , estimated_parameters):
    mean = seriesObj.mean()
    sum_of_squares = 0.0
    d_freedom = len(seriesObj) - estimated_parameters
    for val in seriesObj:
        sum_of_squares += math.pow((val - mean) , 2)
    return (sum_of_squares/d_freedom)
```

```python
#conducting f test to check if the variance is equal or not 
# then conducting t test to check if the distributions are equal or not 
#im doing this to check if the minTemp has any effect on rainTomorrow or not 

#both distribution looks close enough to normal distribution 
sample_var_1 = calculate_variance_of_sample(training_df[training_df['RainTomorrow'] == 'Yes']['MinTemp'].dropna() , 1)
sample_var_2 = calculate_variance_of_sample(training_df[training_df['RainTomorrow'] == 'No']['MinTemp'].dropna() , 1)
sample_mean_1 = training_df[training_df['RainTomorrow'] == 'Yes']['MinTemp'].dropna().mean()
sample_mean_2 = training_df[training_df['RainTomorrow'] == 'Yes']['MinTemp'].dropna().mean()
df1 = len(training_df[training_df['RainTomorrow'] == 'Yes']['MinTemp'].dropna()) - 1
df2 = len(training_df[training_df['RainTomorrow'] == 'No']['MinTemp'].dropna()) - 1
print(StatEssentials.f_test_equal_variance(sample_var_1 , sample_var_2 , df1, df2))
print(StatEssentials.independent_two_sample_t_test_equal_variance(sample_mean_1 , sample_mean_2 , sample_var_1 , sample_var_2 , df1 , df2))
print(sample_mean_1)
print(sample_mean_2)
print(df1)
print(df2)
print(sample_var_1)
print(sample_var_2)
#the t test shows that MinTemp has no relationship with RainTmrw
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'StatEssentials' is your module with the calculate_bin_number_histogram_series_obj function
windGustDirSeries = training_df['WindGustDir'].unique()  # Get unique categories only
list_of_series_objects = []

for val in windGustDirSeries:
    series_obj = training_df[training_df['WindGustDir'] == val]['WindGustSpeed']
    list_of_series_objects.append({'obj': series_obj, 'category': val})

for series_objects in list_of_series_objects:
    print(series_objects['category'])
    sns.histplot(series_objects['obj'], bins=StatEssentials.calculate_bin_number_histogram_series_obj(series_objects['obj']), kde=True)

# Optionally, add plt.show() to display all plots
plt.show()
```

```python
#sample with only yes 

# Assuming 'StatEssentials.calculate_bin_number_histogram_series_obj' is defined
# and 'training_df' is your DataFrame

# Group the DataFrame by 'WindGustDir'
grouped = training_df.groupby('WindGustDir')

# Set up the subplots: adjust rows and cols according to the number of categories
n_categories = len(grouped)
cols = 5  # Number of columns in the subplot grid
rows = (n_categories // cols) + (n_categories % cols)  # Number of rows in the subplot grid


fig, axes = plt.subplots(rows, cols, figsize=(20, 5 * rows))  # Adjust figure size
axes = axes.flatten()  # Flatten in case we have a single row

# Iterate over each group and corresponding subplot axis
temp_list = []
for i, (category, group) in enumerate(grouped):
    wind_gust_speeds = group[group['RainTomorrow'] == 'Yes']['WindGustSpeed'].dropna()  # Drop NaN values
    mean_value = wind_gust_speeds.mean()
    var_value = wind_gust_speeds.var()
    
    temp_list.append({'mean': mean_value, 'var': var_value, 'category': category})
    
    # Plot on the corresponding subplot axis
    sns.histplot(wind_gust_speeds, bins=StatEssentials.calculate_bin_number_histogram_series_obj(wind_gust_speeds), kde=True, ax=axes[i])
    axes[i].set_title(f'Histogram of WindGustSpeed for {category}')
    axes[i].set_xlabel('WindGustSpeed')
    axes[i].set_ylabel('Frequency')

# Hide any unused subplots

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()  # Adjust subplots to fit in the figure area
plt.show()


```

```python
results_df_raining_yes = pd.DataFrame(temp_list)
```

```python
results_df_raining_yes.sort_values(by = 'mean', ascending = False)
```
